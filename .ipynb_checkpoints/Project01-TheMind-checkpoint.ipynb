```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's start by creating a function that generates one sample of the game. Feel free to change this code as you find it suitable, for example by changing how the sampling is created or if you want to sample cards with replacement (this could be useful for the last questions).

```python
def sample_cards(N, n, k):
    """
    Generates a simulated sample of how the cards are shuffled among players. 
    
    Arguments: 
        - N : Total number of cards
        - n : Number of players
        - k : Number of cards per player
        
    Returns: 
        - 2D numpy array where each row represent each player and each column their card in ascending order. 
    """
    
    assert n * k <= N, "There are not enough cards for all the players!"
    
    _cards = np.random.choice(a=np.arange(1, N+1), size=n*k, replace=False)
    _cards = _cards.reshape(n,k)
    _cards.sort(axis=1)
    return _cards
```

Now we can just generate how cards are distributed among players by simply making a call to this function with the desired parameters.

### 1. What is the distribution of the lowest number? 

Repeat the sampling of the cards multiple times for the same choice of `N`, `n` and `k` and plot the distribution of the lowest number in the sample. Use an [histogram](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) to plot the distribution and set the bins to be the available integer number cards. 

```python
# To Do cell
N, n, k, sim = 100, 6, 3, 100

# Create a list contain lowest number in each simulation
min_card_list = []
for sample in range(sim):
    min_card_list.append(np.min(sample_cards(N, n, k)))

# Plot histogram
min_card = np.array(min_card_list)
plt.hist(min_card, bins=N, color='blue', alpha=0.5)
plt.title('Lowest Number Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 2. What is the distribution of the largest gap?

For each simulation, found the largest distance between consecutive numbers in the game. To make this easy, let us first consider the largest gap among the pool of **all** the cards among players (if the largest gap is happening in the hand of one single player, that is ok). This will make the analysis more simple. 

```python
# To Do cell
N, n, k, sim = 100, 6, 3, 100

pool_gap_list = []
for sample in range(sim):
    cards = sample_cards(N, n, k)
    
    # diff is a numpy array consist of difference between consecutive number
    diff = np.diff(cards)
    
    # find the maximum difference for the sample
    pool_gap = np.max(diff)
    pool_gap_list.append(pool_gap)
    
pool_gap = np.array(pool_gap_list)
print(pool_gap)

```

Repeat the experiment but know compute the largest gap between cards of different players. This is the gap that actually matters for the purposes of the game: what is the maximum number of numbers that a player will have to wait before playing the next card?

```python
# To Do cell
def interperson_gap(N, n, k):
    """
    Arguments: 
        - N : Total number of cards
        - n : Number of players
        - k : Number of cards per player
        
    Returns: 
        - The largest gap between cards of different players
    """
    deck = sample_cards(N, n, k)

    # Simulate the first round
    min_card = np.min(deck)
    last_player = np.where(deck == min_card)[0]
    gap_list = []

    # Simulate the playing process assume people can read minds
    for rounds in range(n*k): 
        try:        
            # Remove last players deck from consideration
            available_deck = np.delete(deck, last_player, axis=0)

            # Find the gap and the current minimum
            current_deck = available_deck[available_deck > min_card]
            gap = np.min(current_deck) - min_card
            min_card = np.min(current_deck)

            # Return the row indicies to record player at this round
            last_player = np.where(deck == min_card)[0]
            gap_list.append(gap)
            
        except ValueError:
            # print("Last man standing!")
            break

    return(max(gap_list))

# The maximum result after #sim simulations
print(f"The maximum number of cards a player must wait before playing the next card is {interperson_gap(100, 6, 3)} with k = {3}")
```

How does these compare? Run simulations for different choices of the number of cards and players and see when the difference between these two is larger. 

Compute the mean of the two distributions. When this difference is larger? How does this depends of the total number of players `n`?

#### As k increase, the maximum and and mean waitting time decrease

```python
# To Do cell
import pandas as pd

N, n, sim = 100, 6, 100
k_min, k_max, k_gap_list = 3, 13, []

# Loop through k and simulate
for k in range(k_min, k_max):
    sim_gap_list = []
    for sample in range(sim):
        sim_gap_list.append(interperson_gap(N, n, k))

    sim_gap = np.array(sim_gap_list)
    k_gap_list.append([k, np.max(sim_gap), np.mean(sim_gap)])
    
# Convert the list to NumPy array    
k_gap = np.array(k_gap_list)
column_names = ['k = Hand', 'max_wait', 'mean_wait']

# Convert the NumPy array to a DataFrame with custom column names and index labels
kgap_df = pd.DataFrame(k_gap, columns=column_names)
print(kgap_df)

```

### 3. Deriving distributions

For this section, we are going to use the random sampling generating process from _The Mind_ to investigate and find some familiar distributions.

### 3.1. Binomial Distribution 

Can you think about any way of deriving the binomial distribution? Based on the output of one single game (function `sample_game()`), try to derive something that has Binomial distribution for certain parameter $p$. Run multiple simulations to see this and compare the obtained empirical distribution with the distribution of the binomial distribution (check `numpy.random.binomial` and `scipy.stats.binom`). 

```python
# To Do cell
from scipy.stats import binom
N, n, k, q, numsim= 1000, 60, 4, 3, 40000

# Number of numbers in card that divides q ~ Bin(m, p=1/q)
sim_list = []
for run in range(numsim):
    sim_list.append(np.sum(sample_cards(N, n, k) % q == 0))

# Plot histogram
sim_array = np.array(sim_list)
plt.hist(sim_array, bins=n, color='blue', density=True, alpha=0.5, label="Simulated Distribution")

# Generate and plot Binomial distribution
x = np.arange(min(sim_array), max(sim_array) + 1)
pmf = binom.pmf(x, n*k, 1/q)
plt.plot(x, pmf, 'r-', lw=2, label=f"Bin({n*k}, {1/q:.2f})")

# Labels and title
plt.title('Number of Card Divides q vs. Binomial Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.2. Beta Distribution

The [order statistics](https://en.wikipedia.org/wiki/Order_statistic) of the uniform distribution are such that follow a Beta distribution. If well here we are not exactly sampling real numbers in $[0,1]$ following the uniform distribution, you can imagine that sampling one single number without repetition from $\{ 1, 2, 3, \ldots, N \}$ is not very different...

Compare the distribution of the lowest number in the sample (Part 1) with a [Beta distribution](https://en.wikipedia.org/wiki/Beta_distribution) with adequate choices of parameters. In order to do this, notice that the $k$th order statistics of a sample of $n$ uniformly distributed points in $[0,1]$ is $\text{Beta}(k, n+1-k)$ (These $n$ and $k$ are not the same as before, you need to pick the right values of these).

```python
# To Do cell
from scipy.stats import beta
N, n, k, sim = 1000, 60, 3, 10000

# Create a list contain lowest number in each simulation
min_card_list = []
for sample in range(sim):
    # Scale to unif[0,1]
    scaled_min_card = (np.min(sample_cards(N, n, k)) - 1) / (N - 1)
    min_card_list.append(scaled_min_card)
    
# Plot histogram (density=True to normalize)
min_card = np.array(min_card_list)
plt.hist(min_card, bins=n, color='blue', alpha=0.5, density=True, label="Simulated Distribution")

# Generate and plot Beta distribution
x = np.linspace(0, max(min_card), 1000)
y_values = beta.pdf(x, 1, n * k) # alpha=1, beta=n*k
plt.plot(x, y_values, 'r-', lw=2, label=f"Beta(1, {n * k})")

# Labels and title
plt.title('Lowest Number in the Sample vs. Beta Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.3. Exponential Distribution

If we call $U_1$ the minimum number in the sample (Part 1), then make an histogram with the distribution of the transformed random variable given by 
$$
X = - \ln \left( 1 - \frac{U_1}{N} \right)
$$
and compare this with the distribution of an exponential random variable with parameter $nk$, the total number of cards in the game. 

```python
import math
from scipy.stats import expon

N, n, k, sim = 1000, 60, 3, 50000

# Create a list contain lowest number in each simulation
trans_var_list = []
for sample in range(sim):
    min_card = np.min(sample_cards(N, n, k))
    trans_var = - math.log(1 - (min_card / N))
    trans_var_list.append(trans_var)

# Plot histogram
trans_var = np.array(trans_var_list)
plt.hist(trans_var, bins=n, color='blue', alpha=0.5, density=True, label="Simulated Distribution")

# Generate and plot Exponential distribution
x = np.linspace(0, max(trans_var), 1000)
scale = 1 / (n * k)
y_values = expon.pdf(x, scale=scale)
plt.plot(x, y_values, 'r-', lw=2, label=f"exp(1/{n * k})")

# Labels and title
plt.title('Transformed R.V. in the Sample vs. Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

### 4. What happens when $n * k << N$?

Some of the distribution we saw in previous sections are just approximate. Study what happens when we increase the number of cards. Try to understand how the cases of sampling with and without reposition compare as $n * k << N$. What do you expect to see in these cases? When the difference between the two sampling strategies will lead to similar or different results. 

#### Simulation Result
At first glance, sampling with without replacement seems to generate the same result if the card number is large enough. They both still look like exponential random variable, and the only difference is the scale. This is likely due to the memoryless property of exponential random variable. More rigorous math calculation needed in order to verify the result. Other sampling method may not generate the same result.

```python
# To Do cell
N, n, k= 10000, 6, 3
sim = int(N / (n*k)) - 1

# Sample without replacement
min_card_list = []
for sample in range(sim):
    min_card = np.min(sample_cards(N, n, k))
    min_card_list.append(min_card)

# Plot Histogram
min_card = np.array(min_card_list)
plt.hist(min_card, bins=n*k*5, color='blue', density=False, alpha=0.5, label="Minimum Card Sample without Replacement")

# Sample with replacement
cards = np.arange(1, N+1)
min_card_list2 = []

for sample in range(sim):
    # Skip the reshape procedure
    card_sample = np.random.choice(cards, size = n*k, replace = False)
    
    # Create a boolean mask indicate if sample in cards, idea from chatGPT
    mask = np.isin(cards, card_sample)
    
    # Remove card_sample from cards
    cards = cards[~mask]
    min_card_list2.append(np.min(card_sample))

# Plot histogram
min_card2 = np.array(min_card_list2)
plt.hist(min_card2, bins=n*k*5, color='yellow', density=False, alpha=0.5, label="Minimum Card Sample with Replacement")

plt.title('Sample with and without Replacement')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
